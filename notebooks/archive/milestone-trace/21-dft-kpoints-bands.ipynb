# k-Points And Band Diagnostics

This notebook runs a non-SCF band path on top of a converged toy density. Γ-point behavior remains the default; the path shows how `0.5|G+k|²` enters the operator.

```python
from mlx_atomistic.dft import BandPath, DFTSystem, KPoint, KPointMesh, SCFConfig, run_band_structure, run_scf

system = DFTSystem.one_center(grid_shape=(4, 4, 4))
scf = run_scf(system, config=SCFConfig(max_iterations=1, solver="dense", seed=5))
path = BandPath([KPoint.gamma(), KPoint((0.25, 0.0, 0.0), label="X")])
bands = run_band_structure(system, scf, path, n_bands=1)
{
    "gamma_weight": KPointMesh.gamma().points[0].weight,
    "band_shape": bands.eigenvalues.shape,
    "reused_density": bands.reused_density,
}
```
