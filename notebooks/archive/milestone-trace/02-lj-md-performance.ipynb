# Lennard-Jones MD Performance

Compare all-pairs, static-neighbor, dynamic-neighbor, and Langevin NVT dynamic-neighbor paths.

```python
import matplotlib.pyplot as plt

from mlx_atomistic.benchmarks.lj_md import run_case

particle_counts = [32, 64, 128]
modes = ("all-pairs", "static-neighbor", "dynamic-neighbor", "nvt-dynamic-neighbor")
results = []
for particles in particle_counts:
    for mode in modes:
        results.append(
            run_case(
                particles=particles,
                steps=10,
                mode=mode,
                density=0.8,
                temperature=1.0,
                seed=7,
            )
        )

results
```

```python
plt.figure(figsize=(7, 4))
for mode in modes:
    xs = [result.particles for result in results if result.mode == mode]
    ys = [result.ms_per_step for result in results if result.mode == mode]
    plt.plot(xs, ys, marker="o", label=mode)

plt.xlabel("particles")
plt.ylabel("ms / step")
plt.legend()
plt.tight_layout()
```
