# Reference Validation

This notebook demonstrates the static reference-comparison layer. QE/CP2K are not executed in CI; reference fixtures are stored as data and compared with explicit tolerances.

```python
from mlx_atomistic.dft import DFTSystem, ReferenceDFTCase, SCFConfig, compare_reference_case, run_scf

system = DFTSystem.one_center(grid_shape=(4, 4, 4))
result = run_scf(system, config=SCFConfig(max_iterations=1, solver="dense", seed=13))
case = ReferenceDFTCase("toy-one-center", "static-fixture", result.total_energy, 1e-8)
comparison = compare_reference_case(case, observed_energy=result.total_energy)
comparison.to_dict()
```
