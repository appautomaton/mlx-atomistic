# Davidson Solver Diagnostics

This notebook compares the Davidson public solver boundary with the dense reference on a tiny grid. Dense remains the correctness oracle for small cases.

```python
import numpy as np

from mlx_atomistic.dft import DFTSystem, DavidsonDiagonalizer, DenseHamiltonianReference, KohnShamOperator, SCFConfig, run_scf

system = DFTSystem.one_center(grid_shape=(4, 4, 4))
scf = run_scf(system, config=SCFConfig(max_iterations=1, solver="dense", seed=5))
operator = KohnShamOperator.from_density(system.grid, system.pseudopotential.field(system.grid), scf.density)
dense = DenseHamiltonianReference(operator).diagonalize(1)
davidson = DavidsonDiagonalizer().solve(operator, n_orbitals=1)
{
    "dense_eigenvalue": float(np.array(dense.eigenvalues)[0]),
    "davidson_eigenvalue": float(np.array(davidson.eigenvalues)[0]),
    "max_residual": float(np.max(np.array(davidson.residuals))),
    "metadata": davidson.metadata,
}
```
