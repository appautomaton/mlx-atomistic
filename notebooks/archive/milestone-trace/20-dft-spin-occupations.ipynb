# Spin And Occupations

This notebook exercises the collinear spin-density helpers and Fermi-Dirac occupation model. It is a diagnostics layer first; full spin-polarized SCF accuracy remains bounded by the prototype XC model.

```python
import numpy as np

from mlx_atomistic.dft import DFTSystem, FermiDiracOccupations, SCFConfig, magnetization_density, run_scf, spin_density_from_orbitals

system = DFTSystem.one_center(grid_shape=(4, 4, 4))
result = run_scf(system, config=SCFConfig(max_iterations=1, solver="dense", seed=9))
occupations = FermiDiracOccupations(2.0, temperature=0.05).resolve(result.orbital_eigenvalues)
rho_up, rho_down = spin_density_from_orbitals(result.orbitals, result.orbitals, system.grid, up_occupations=[1.0], down_occupations=[1.0])
magnetization = magnetization_density(rho_up, rho_down)
{
    "electron_count": occupations.electron_count,
    "chemical_potential": occupations.chemical_potential,
    "magnetization_integral": float(np.sum(np.array(magnetization)) * system.grid.dv),
}
```
