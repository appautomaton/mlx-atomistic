# Nonlocal Projector Inspection

This notebook checks the separable nonlocal pseudopotential path. UPF metadata is converted into normalized real-space projectors, then applied inside the Kohn-Sham operator.

```python
from pathlib import Path

from mlx_atomistic.dft import DFTSystem, Ion, IonCollection, NonlocalPseudopotentialOperator, SCFConfig, read_upf, run_scf

ROOT = Path.cwd().parent if Path.cwd().name == "notebooks" else Path.cwd()
upf = read_upf(ROOT / "vendors/quantum-espresso/pseudo/Si_r.upf")
system = DFTSystem(cell=(8, 8, 8), grid_shape=(4, 4, 4), ions=IonCollection([Ion("Si", (4, 4, 4), upf)]))
result = run_scf(system, config=SCFConfig(max_iterations=1, solver="dense", seed=3))
operator = NonlocalPseudopotentialOperator.from_ions(system.ions, system.grid)
{
    "nonlocal_applied": result.nonlocal_applied,
    "projector_count": operator.projectors.count,
    "nonlocal_energy": result.energy_by_term["nonlocal_pseudopotential"],
}
```
