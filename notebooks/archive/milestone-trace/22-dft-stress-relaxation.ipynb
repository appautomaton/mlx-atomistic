# Stress And Relaxation Diagnostics

This notebook computes a diagonal finite-difference stress estimate and one ion-relaxation smoke step. Stress is orthorhombic only in this milestone.

```python
from mlx_atomistic.dft import GeometryOptimizationConfig, SCFConfig, finite_difference_stress, geometry_demo_system, optimize_geometry

system = geometry_demo_system("gaussian-dimer", grid_shape=(4, 4, 4))
config = SCFConfig(max_iterations=20, solver="dense", seed=17, convergence_mode="either")
relax = optimize_geometry(system, config=GeometryOptimizationConfig(max_steps=1, scf_config=config))
stress = finite_difference_stress(system, config=SCFConfig(max_iterations=1, solver="dense", seed=17))
{
    "relax_status": relax.status,
    "steps": len(relax.steps),
    "stress": stress.to_dict(),
}
```
